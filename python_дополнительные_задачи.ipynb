````python
# написать функции, возращающую значение синуса для списка элементов
import math
import numpy as np

# объявление функции
def get_sin(lst_values:list) -> list:
    """
    Функция возвращает значение синуса каждого элемента списка.

    Args:
    - lst_values -- список вещественных чисел
    """

    lst_return = list()
    for value in lst_values:
        lst_return.append(np.round(math.sin(value), 2))
    return lst_return
````

````python
# вызов функции
# задаем список углов
l = [0, math.pi/6, math.pi/4, math.pi/3, math.pi/2]
get_sin(tuple(l))
````
